```python
# fix random seed for reproducibility
seed = 8651
np.random.seed(seed)
```

### Load the Dataset

```python
df = pd.read_table('../datasets/wine.tab')
```

```python
df.head()
```

```python
df.drop(df.index[[0, 1]], inplace=True)
df.reset_index(drop=True)
```

```python
X_data = df.drop('Wine', axis=1)
y_data = df['Wine']
```
